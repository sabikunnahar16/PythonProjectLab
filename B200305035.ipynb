```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv('/content/FuelConsumptionCo2 - FuelConsumptionCo2.csv')
```

```python
x = df[['FUELCONSUMPTION_COMB_MPG']].values
y= df['CO2EMISSIONS'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train_b = np.c_[np.ones((x_train.shape[0], 1)), x_train]
x_test_b = np.c_[np.ones((x_test.shape[0], 1)), x_test]
```

```python
theta = np.linalg.inv(x_train_b.T.dot(x_train_b)).dot(x_train_b.T).dot(y_train)
c= theta[0]
m= theta[1]
print(c)
print(m)
```

```python
print("model y=",m,"x +",c)
```

```python
y_pred = x_test_b.dot(theta)
print(y_pred)
```

```python
mean_val= mean_squared_error(y_test, y_pred)
print("Mean Squared Error=",mean_val)
```

```python
plt.figure(figsize=(10,8))
plt.title("FUELCONSUMPTION_COMB_MPG vs CO2 Emissions")
plt.xlabel("FUELCONSUMPTION_COMB_MPG")
plt.ylabel("CO2 Emissions")
plt.scatter(x_train, y_train, color="green", label='Training Data',alpha=0.6)
plt.scatter(x_test, y_test, color="blue",label='Testing Data',alpha=0.6)
plt.plot(x_test, y_pred, color="red", linewidth=2, label='Predicted line')
plt.legend()
plt.grid(True)
plt.show()
```
