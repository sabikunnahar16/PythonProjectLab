```python
import pandas as pd
df=pd.read_csv('/content/Fuel_Consumption_2000-2022.csv')
print(df)
# Display the first 5 rows (default)
print("First 5 rows of the DataFrame:")
print(df.head())

# Display the last 5 rows (default)
print("\nLast 5 rows of the DataFrame:")
print(df.tail())

# Generate descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Provide concise summary of the DataFrame
print("\nInfo about the DataFrame:")
print(df.info())

# Compute mean, median, min, max
print("\nMean of numeric columns:")
print(df.mean(numeric_only=True))

print("\nMedian of numeric columns:")
print(df.median(numeric_only=True))

print("\nMinimum values in each column:")
print(df.min(numeric_only=True))

print("\nMaximum values in each column:")
print(df.max(numeric_only=True))

# Compute sum, count, standard deviation, variance
print("\nSum of numeric columns:")
print(df.sum(numeric_only=True))

print("\nCount of numeric columns:")
print(df.count(numeric_only=True))

print("\nStandard deviation of numeric columns:")
print(df.std(numeric_only=True))

print("\nVariance of numeric columns:")
print(df.var(numeric_only=True))

# Fill missing values with 0
df_filled = df.fillna(0)
print("\nDataFrame after filling NaN values with 0:")
print(df_filled.head())

# Drop rows with missing values
df_dropped_rows = df.dropna(axis=0, how='any')
print("\nDataFrame after dropping rows with NaN values:")
print(df_dropped_rows.head())

# Drop columns with missing values
df_dropped_columns = df.dropna(axis=1, how='any')
print("\nDataFrame after dropping columns with NaN values:")
print(df_dropped_columns.head())

# Sort the DataFrame by a specific column in descending order (change column name accordingly)
sorted_df = df.sort_values(by='CO2 Emissions(g/km)', ascending=False)
print("\nDataFrame sorted by CO2 Emissions in descending order:")
print(sorted_df.head())

# Group by a specific column (change column name accordingly)
grouped_df = df.groupby('Fuel Type').mean(numeric_only=True)
print("\nGrouped DataFrame by Fuel Type with mean values:")
print(grouped_df)

# Create a pivot table
pivot = df.pivot_table(values='CO2 Emissions(g/km)', index='Fuel Type', columns='Cylinders', aggfunc='mean')
print("\nPivot Table:")
print(pivot)

# Apply a function (Example: Convert CO2 emissions to pounds per mile)
df['CO2 Pounds/mile'] = df['CO2 Emissions(g/km)'].apply(lambda x: x * 0.00220462)
print("\nDataFrame after applying function to convert CO2 emissions to pounds per mile:")
print(df.head())
```
